```python
import requests

url_station_status = 'https://velib-metropole-opendata.smovengo.cloud/opendata/Velib_Metropole/station_status.json'

def fetch_station_status():
        url=url_station_status
        headers = {'Content-Type': 'application/json'}

        # call the api velib to get the station status
        response = requests.get(url=url, headers=headers)
        data = response.json()
        return data

raw_data = fetch_station_status()
print(raw_data)
```

```python
### getting the data of stations
raw_stations = raw_data['data']['stations']
```

```python
import datetime
import pytz

last_updated_other = raw_data['lastUpdatedOther']
heure_posix = last_updated_other
heure_lisible = datetime.datetime.fromtimestamp(heure_posix).astimezone(pytz.timezone('Europe/Paris')).strftime('%Y-%d-%m %H:%M:%S')

print(heure_lisible)
```

```python
import pandas as pd

# selecting only the necessary station properties for the final dataframe
df_columns = ['station_id', 'num_bikes_available', 'num_bikes_mechanical', 'num_bikes_ebike', 'num_docks_available', 'is_renting', 'last_updated_readable', 'last_updated_other', 'last_reported_readable', 'last_reported']
# creating the list of stations
stations_list = []

for raw_station in raw_stations:
    station_list = []

    # add all the necessary stations into the station list
    station_list.append(raw_station['station_id'])
    station_list.append(raw_station['num_bikes_available'])
    station_list.append(raw_station['num_bikes_available_types'][0]['mechanical'])
    station_list.append(raw_station['num_bikes_available_types'][1]['ebike'])
    station_list.append(raw_station['num_docks_available'])
    station_list.append(raw_station['is_renting'])
    station_list.append(heure_lisible)
    station_list.append(heure_posix)

    # converting last reported in posix to readable time for humans
    last_reported = raw_station['last_reported']
    last_reported_readable = datetime.datetime.fromtimestamp(last_reported).astimezone(pytz.timezone('Europe/Paris')).strftime('%Y-%d-%m %H:%M:%S')
    
    station_list.append(last_reported_readable)
    station_list.append(last_reported)

    # add the station t the global stations list
    stations_list.append(station_list)

print(stations_list)
```

```python
df_stations = pd.DataFrame(data=stations_list, columns=df_columns)
df_stations.head()
```
